### Домашнее задание 1
#### 1.1. Скачать данные по ссылке

```python
import pandas as pd
df = pd.read_csv(r"kc-house-data.csv", sep = ',',encoding='windows-1251')
df['price'] = df['price'].astype (int)
df.head(5)
```

#### 1.2 Изучите стоимости недвижимости


```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16, 4))
plt.hist(df['price'])
plt.title("График стоимости недвижимости")
plt.xlabel("Цена")
plt.ylabel("Количество")
```

##### Вывод: больше всего домов со стоимостью до 800 000.

#### 1.3. Изучите распределение квадратуры жилой

```python
plt.figure(figsize=(8, 4))
plt.hist(df["sqft_living"])
plt.title("График распределение квадратуры жилой")
plt.xlabel("Жилая квадратура")
plt.ylabel("Количество")
```

##### Вывод: больше всего домов с квадратурой живой от 2000 до 3000 м2.

#### 1.4. Изучите распределение года постройки

```python
plt.figure(figsize=(8, 4))
plt.hist(df["yr_built"])
plt.title("График распределение года постройки")
plt.xlabel("Год постройки")
plt.ylabel("Количество")
```

##### Вывод: основная масса домов относительно новые.

### Домашнее задание 2
#### 2.1 Изучите распределение домов от наличия вида на набережную


```python
data=df['waterfront'].value_counts()
plt.pie(data.values,autopct='%1.1f%%',labels=data.index)
plt.title("График распределение домов от наличия вида на набережную")
plt.legend(['нет вида','есть вид'])
```

##### Вывод: свыше 99% домов не имеют вида на набережную.

### 2.2 Изучите распределение этажей домов

```python
data2=df['floors'].value_counts()
plt.bar(data2.index,data2.values)
plt.ylabel("Количество домов")
plt.xlabel("Этажность дома")
```

##### Вывод: больше всего (свыше 10000) одноэтажных домов.

#### 2.3 Изучите распределение состояния домов

```python
data2=df['condition'].value_counts()
plt.bar(data2.index,data2.values)
plt.ylabel("Количество домов")
plt.xlabel("Состояние домов по бальной системе")
```

##### Вывод: больше всего домов(около 14000) с состоянием по бальной шкале под номером 3.

##### Давайте построим матрицу карелляции по двум признакам "цене" и "количеству спален".
##### Как видно из графика зависимость цены от количества спеален есть но она не такая сильная
